```python
import glob
import pandas
import splat
import splat.plot as splot

#change to your respective paths
quasar_templates_path = "C:/Users/brend/OneDrive/Documents/UCSD/Burgasser_Research"
```

```python
files = glob.glob(quasar_templates_path + '/*.txt')
quasar_templates = {}

for f in files:
    file_name = f.split('_')[-1].replace('.txt', '')   #get the identifying part of the file name
    df = pandas.read_fwf(f, names = ['wave', 'flux', 'noise'])
    sp = splat.Spectrum(wave = df['wave'], flux = df['flux'], name = file_name)
    quasar_templates[file_name] = sp
    
quasar_templates
```

```python
keys_list = list(quasar_templates.keys())
quasar_list = [quasar_templates[k] for k in keys_list]

splot.plotSpectrum(quasar_list, colormap = 'viridis')
```

```python
spectra = splat.Spectrum('example.fits')
splat.classifyByTemplate(spectra, quasar_templates, plot = True)
```
